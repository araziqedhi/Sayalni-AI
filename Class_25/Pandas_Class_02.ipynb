```python
def name_generator(max_len):
    r=np.random.randint(3,max_len,1)[0]
    fc=chr(np.random.randint(65,90,1)[0])
    remaining_letters=np.random.randint(97,122,r)
    remaining_letters=list(map(chr,remaining_letters))
    return (fc+"".join(remaining_letters))
    
list(map(name_generator,np.random.randint(4,20,100)))
```

```python
def dob_generator(start,end,month_start,month_end,day_start,day_end):
    year = np.random.randint(start,end,1)[0]
    month = np.random.randint(month_start,month_end,1)[0]
    day = np.random.randint(day_start,day_end,1)[0]
    return pd.to_datetime(f"{year}/{month}/{day}")
dob_generator(1970,2000,1,12,1,30)
```

```python
sample=10100
df = pd.DataFrame({"roll no":np.arange(1,sample+1),
                  "Name":list(map(name_generator,np.random.randint(4,20,sample))),
                   "Father Name":list(map(name_generator,np.random.randint(4,20,sample))),
                   "Course":np.random.choice(["AI","Web","ChatBot"],sample),
                   "Fee":np.random.choice([1000,2000,3000],sample),
                   "Date of Birth": list(map(lambda _: dob_generator(1970, 2000, 1, 12,1,29), np.random.randint(4, 20, sample)))
                  })
df
df1=pd.DataFrame(np.random.randint(1,100,size=(sample,5)),
                columns=["s1","s2","s3","s4","s5"])

df=pd.concat([df,df1],axis=1)
df
```

## How to apply Custom Function on DataFrame

```python
def grade(per):
    grade=None 
    if per >=80:
        grade ="A"
    elif per >=70:
        grade="B"
    elif per >=60:
        grade = "C"
    else :
        grade = "F"
    return grade
```

```python
df['grade']=df["percentage"].apply(grade)
```
