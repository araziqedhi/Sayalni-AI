## Now Open File With Block

```python
with open ("./text.txt") as f:
    print(f.read())
```

```python
with open ("./text.txt") as f:
    f.seek(0)
    print(f.read())
```

```python
with open ("./text.txt") as f:
    f.seek(1)
    print(f.read())
```

```python
with open ("./text.txt") as f:
    f.seek(2)
    print(f.read())
```

```python
with open ("./text.txt",mode="w") as f:
    f.write("Hello World")
```

```python
with open ("./text.txt",mode="w+") as f:
    f.write("Hello Workd")
    f.seek(0)
    print(f.read())
```

```python
with open ("./text.txt",mode="r+") as f:
    f.write("Hello Workd")
    print(f.read())
```

```python
with open ("./text.txt",mode="r+") as f:
    f.seek(len(f.read())+1)
    f.write("Hello Workd")
    f.seek(0)
    print(f.read())
    
```

```python
with open ("./text.txt",mode="rb") as f:
    print(type(f))
    print(f.read())
```

## Store Input User in File Using While Loop

```python
while True:
    name=input("Enter your name ")
    if name.lower()=="exit" or name.lower()=="x":
        break
    with open ("studentName.txt","a") as f: # Create new file 
        f.write(name+"\n")
```
