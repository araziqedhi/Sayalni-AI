```python
for rid,row in enumerate(l):
    print(rid,row)
    for cid,_ in enumerate(row):
        print(cid,end="***")
    print()
```

# Real World DAta 
 * scalar 
 * vector
 * matrics
 * n-dimension array

```python
## 1) Scalar
l1=np.asarray(7)
display(l1)
display(type(l1))
l1=np.array(l1)
display(l1)
display(type(l1))

print("dimesion:",l1.ndim)
print("size:",l1.size)
print("no access in each dimesion ",l1.shape)
print("data type",l1.dtype)
```

```python
## 1) Vector
l1=np.asarray([1,2,3])
display(l1)
display(type(l1))
l1=np.array(l1)
display(l1)
display(type(l1))

print("dimesion:",l1.ndim)
print("size:",l1.size)
print("no access in each dimesion ",l1.shape)
print("data type",l1.dtype)
```

```python
## 1) Matrix
l1=np.asarray([[1,2,3]])
display(l1)
display(type(l1))
l1=np.array(l1)
display(l1)
display(type(l1))

print("dimesion:",l1.ndim)
print("size:",l1.size)
print("no access in each dimesion ",l1.shape)
print("data type",l1.dtype)
```
