## np.in1d

```python
ubl = np.arange(1,101)
display(ubl)
sbp = np.array([1000,70,5,40000,7899])
selected_list=[]
for cnic in sbp:
    if cnic in ubl:
        print(True,cnic)
        selected_list.append(cnic)
selected_list
  
```

```python
## same process with np.in1d
```

```python
ubl = np.arange(1,101)
display(ubl)
sbp = np.array([1000,70,5,40000,7899])

np.in1d(sbp,ubl)
```

```python
ubl = np.arange(1,101)
display(ubl)
sbp = np.array([1000,70,5,40000,7899])

sbp[np.in1d(sbp,ubl)]
```

```python
ubl = np.arange(1,101)
display(ubl)
sbp = np.array([1000,70,5,40000,7899])

np.intersect1d(sbp,ubl)
```
