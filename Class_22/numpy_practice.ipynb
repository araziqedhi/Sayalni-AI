```python
import numpy as np
import cv2
```

```python
image1 = cv2.imread('./images.jpg')
image2 = cv2.imread('./istockphoto-1157655660-612x612.jpg')

image1 = cv2.resize(image1, (1000, 1000))
image2 = cv2.resize(image2, (1000, 1000))
result = np.add(image1, image2)
```

```python
cv2.imshow('Result', result)
```

```python
print(result)
```

```python
cv2.imshow('Result', result)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imwrite('result.jpg', result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

image1 = np.array(Image.open('./images.jpg'))
image2 = np.array(Image.open('./istockphoto-1157655660-612x612.jpg'))
width, height = 1000, 1000
image1 = np.resize(image1, (height, width))
image2 = np.resize(image2, (height, width))
result = np.add(image1, image2)
plt.imshow(result)
plt.axis('off')
plt.show()

result_image = Image.fromarray(result.astype(np.uint8))
result_image.save('result.jpg')
```
