## Opening and Analyzing Youtube with help of urllib and beautiful Soup

```python
text = urllib.request.urlopen("https://www.youtube.com/watch?v=fx60nv2CZ5Q")
text
soup = BeautifulSoup(text, 'html.parser')

print(soup.prettify())

tags=re.findall("<.*?>",soup)
tags
```
